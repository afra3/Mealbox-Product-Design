# 2. Run regression on filtered data

```python
dt = pd.read_csv("/Users/afraaa/Desktop/data.csv")
```

```python
dt.columns
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
#Build partworth table for individuals
pw_ID_c = pd.DataFrame(columns=['Vegetarian', '30 min', '60 min','Bio', 'Delivered before each meal', '6.4', '9.3'])
pw_ID_i = pd.DataFrame(columns=['intercept'])
```

```python
for i in range(19):
    temp = dt[dt['ID']==i+1]
    X = temp[['Vegetarian', '30 min', '60 min','Bio', 'Delivered before each meal', '6.4', '9.3']]
    y = temp[['rating']]
    reg = LinearRegression().fit(X, y)
    pw_ID_c.loc[i] = reg.coef_[0]
    pw_ID_i.loc[i] = reg.intercept_
```

```python
temp = dt
X = temp[['Vegetarian', '30 min', '60 min','Bio', 'Delivered before each meal', '6.4', '9.3']]
y = temp[['rating']]
reg = LinearRegression().fit(X, y)
pw_ID_c.loc['All'] = reg.coef_[0]
pw_ID_i.loc['All'] = reg.intercept_
```

```python
pw_ID_c.to_csv("/Users/afraaa/Desktop/pw_ID_c.csv")
pw_ID_i.to_csv("/Users/afraaa/Desktop/pw_ID_i.csv")
```

```python
#Build partworth table for segments
pw_seg_c = pd.DataFrame(columns=['Vegetarian', '30 min', '60 min','Bio', 'Delivered before each meal', '6.4', '9.3'])
pw_seg_i = pd.DataFrame(columns=['intercept'])
```

```python
for j in ['gender', 'age','geography', 'location', 'frequency']:
    for i in set(dt[j]):
        temp = dt[dt[j]==i]
        X = temp[['Vegetarian', '30 min', '60 min','Bio', 'Delivered before each meal', '6.4', '9.3']]
        y = temp[['rating']]
        reg = LinearRegression().fit(X, y)
        pw_seg_c.loc[i] = reg.coef_[0]
        pw_seg_i.loc[i] = reg.intercept_
```

```python
temp = dt
X = temp[['Vegetarian', '30 min', '60 min','Bio', 'Delivered before each meal', '6.4', '9.3']]
y = temp[['rating']]
reg = LinearRegression().fit(X, y)
pw_seg_c.loc['All'] = reg.coef_[0]
pw_seg_i.loc['All'] = reg.intercept_
```

```python
pw_seg_c.to_csv("/Users/afraaa/Desktop/pw_seg_c.csv")
pw_seg_i.to_csv("/Users/afraaa/Desktop/pw_seg_i.csv")
```

```python
imp_seg = pd.DataFrame(columns=['Meal type', 'Cook time', 'Wellness','Delivery mode','Price'])
```

```python
for i in pw_seg_c.index:
    m=pw_seg_c.loc[i]
    r=[]
    r.append(max(0,m['Vegetarian'])-min(0,m['Vegetarian']))
    r.append(max(0,m['30 min'],m['60 min'])-min(0,m['30 min'],m['60 min']))
    r.append(max(0,m['Bio'])-min(0,m['Bio']))
    r.append(max(0,m['Delivered before each meal'])-min(0,m['Delivered before each meal']))
    r.append(max(0,m['6.4'],m['9.3'])-min(0,m['6.4'],m['9.3']))
    imp_seg.loc[i] = r/sum(r)
```
